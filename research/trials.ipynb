### Download data from Google Drive

```python
import gdown
```

```python
url = " https://drive.google.com/file/d/1rU2T1zDALwO-QyC8cVPtJrk0lCRLOal9/view?usp=sharing"
file_id = url.split('/')[-2]
```

```python
prefix = "https://drive.google.com/uc?/export=download&id="
gdown.download(prefix+file_id,"artifacts.zip ")
```
